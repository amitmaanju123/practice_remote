# 1. Arithmetic Operators

```python
a=10
b=3

print(a + b)   
print(a - b)   
print(a * b)   
print(a / b)   
print(a // b)  # Floor Division
print(a % b)   # Modulus (remainder)
print(a ** b)  # Exponentiation (power)
```

# 2. Comparison (Relational) Operators

```python
a = 10
b = 5

print(a == b)  
print(a != b)  
print(a > b)   
print(a < b)   
print(a >= b)  
print(a <= b)  
```

# 3. Assignment Operators

```python
x = 10
x += 5
print(x)  

x -= 3
print(x)  

x *= 2
print(x)  

x /= 4
print(x)  
```

# 4. Logical Operators

```python
a = 10
b = 5

print(a > 5 and b < 10)  
print(a > 5 or b > 10)   
print(not(a > 5)) 
```

# 5. Bitwise Operators

```python
a = 5   
b = 3   

print(a & b)   
print(a | b)   
print(a ^ b)   
```

# 6. Membership Operators

```python
fruits = ['apple', 'banana', 'mango']

print('apple' in fruits)      
print('grape' not in fruits)  
```

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is z)      
print(x is y)       
print(x == y)      
print(x is not y)   
```

# copy

```python
#copy
A shallow copy creates a new list object, 
but the elements themselves are references to the same objects in the original list.
```

```python
#Using copy module
import copy
copy3 = copy.copy(l)
```

```python
#in 1d
l=[1,2,3,4,5]
import copy
l1 = copy.copy(l)
l1[2]=20
print(l1)
print(l)
```

```python
#in case of 2d
it create a new outer list. But the inner lists are still the same objects:
```

```python
l = [[1, 2], [3, 4]]
import copy
l1 = copy.copy(l)
l1[0][0] = 100
print(l1)
print(l)
#print(id(l1[0]))
#print(id(l[0]))
```

```python
A deep copy duplicates all nested objects, 
so changes in the original list do not affect the copy.
```

```python
import copy
l = [[1, 2], [3, 4]]
l1= copy.deepcopy(l)
l[0][0] = 500
print(l1) 
print(l)
```
